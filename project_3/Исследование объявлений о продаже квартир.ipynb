По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**

- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)

- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

Пропуски в этих признаках имеет смысл рассмотреть вместе.Как правило площадь кухни и жилая зависият от общей площади квартиры. Посмотрим еще раз на долю пропусков у этих признаков.Природа их возникновения возможно кроется в том,что люди не указывали в объявлениях эти параметры.

```python
df[['living_area','kitchen_area']].isna().mean() # доля пропусков в признках living_area и kitchen_area
```

Получилось, что из 8% и 9.6% соответственно. Это достаточно много и имеет смысл их заполнить.Прсмотрим с каким из признаков есть наиболее теснаяя связь.

```python
df_correlation = df.corr()
df_correlation.sort_values(by='kitchen_area',ascending=False)['kitchen_area'].reset_index().head()
```

```python
df_correlation.sort_values(by='living_area',ascending=False)['living_area'].reset_index().head()
```

Действительно, как и предполагалось это действительно общая площадь.Хотя надо заметиить, что площадь кухни не так сильно зависит от общей площади квартиры.Таким образом эти пропуски можно заменить с учетом коэффициента соотношения общей площади к жилой/кухни соответственно.

```python
living_rat = df['living_area'].mean() / df['total_area'].mean()
kitchen_rat = df['kitchen_area'].mean() / df['total_area'].mean()
df['living_area'].fillna(living_rat * df['total_area'],inplace=True)
df['kitchen_area'].fillna(kitchen_rat * df['total_area'],inplace=True)
```

```python
df[['living_area','kitchen_area']].isna().mean()
```
